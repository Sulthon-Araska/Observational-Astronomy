```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.time import Time
from astroplan import FixedTarget, Observer
from astropy.coordinates import SkyCoord, EarthLocation
from astroplan.plots import dark_style_sheet, plot_altitude, plot_sky
from pytz import timezone

# Observation location
location = EarthLocation.from_geodetic(107.615556, -6.824444, 1310, ellipsoid='WGS84')
observer = Observer(name='Bosscha Observatory', location=location, timezone=timezone('Asia/Jakarta'))

# Observation time
start_time   = Time(['2023-10-20 12:00:00'])
end_time     = Time(['2023-10-20 21:00:00'])
delta_t      = end_time - start_time
observe_time = start_time + delta_t*np.linspace(0, 1, 75)
```

# **Visibility & Sky Charts**

```python
# Objects
SX_Phe    = FixedTarget(name='SX Phe',    coord=SkyCoord('23:46:32.89','-41:34:54.8', unit=(u.hourangle, u.deg), frame='icrs'))
V1889_Aql = FixedTarget(name='V1889 Aql', coord=SkyCoord('20:11:38.88','+16:06:27.9', unit=(u.hourangle, u.deg), frame='icrs'))
LS_Cet    = FixedTarget(name='LS Cet',    coord=SkyCoord('02:24:29.73','+08:12:30.7', unit=(u.hourangle, u.deg), frame='icrs'))
γ_Equ     = FixedTarget(name='γ Equ',     coord=SkyCoord('21:10:20.50','+10:07:53.7', unit=(u.hourangle, u.deg), frame='icrs'))
Moon      = FixedTarget(name='Moon',      coord=SkyCoord('06:16:16.00','+27:50:55.0', unit=(u.hourangle, u.deg), frame='icrs'))
objects   = [SX_Phe, V1889_Aql, LS_Cet, γ_Equ, Moon]

# Altitude Plot
plot_altitude(objects, observer, observe_time, style_sheet=dark_style_sheet, brightness_shading=True)
plt.legend(shadow=True, loc=2)
plt.tight_layout()
plt.show()

#Azimuth Plot
plot_sky(SX_Phe,    observer, observe_time)
plot_sky(V1889_Aql, observer, observe_time)
plot_sky(LS_Cet,    observer, observe_time)
plot_sky(γ_Equ,     observer, observe_time)
plot_sky(Moon,      observer, observe_time)
plt.legend(shadow=True, loc=2)
plt.tight_layout()
plt.show()
```
